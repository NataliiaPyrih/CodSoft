```python
from scipy.stats import ks_2samp
flat_sample_data = [item for sublist in df_sample for item in sublist]
flat_population_data = [item for sublist in df for item in sublist]
statistic, p_value = ks_2samp(flat_sample_data, flat_population_data)
if(p_value >= 0.05):
  print(f"Sample is represantative: {p_value}")
else:
  print(f"Sample is not represantative: {p_value}")
```

##Machine Learning

```python
from sklearn.model_selection import train_test_split, KFold,  cross_val_score
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
X=df_sample.drop(['Class'], axis=1).values
y=df_sample['Class'].values.ravel()
X_train, X_test, y_train, y_test= train_test_split(X, y, stratify=y, test_size=0.2, random_state=13)
```

```python
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
models={"Logistic Regression": LogisticRegression(solver='liblinear', max_iter=2000),
        "Support Vector machines":SVC(),
        "K-Nearest Neighbors":KNeighborsClassifier(),
        "Decision Tree": DecisionTreeClassifier()}
results={}
for name, model in models.items():
 kFold = KFold(n_splits=3, random_state=13, shuffle=True)
 result = cross_val_score(model, X_train_scaled, y_train, cv=kFold, scoring='accuracy')
 results[name]=result
```

```python
for name, score in results.items():
  print(f'{name} accuracy: {score.mean()}')
```

```python
names = list(results.keys())
accuracy_values = list(results.values())
plt.boxplot(accuracy_values)
plt.show()
```

```python
predictions={}
accuracy_scores={}
for name, model in models.items():
 model.fit(X_train_scaled, y_train)
 y_pred=model.predict(X_test_scaled)
 predictions[name]=y_pred
 accuracy_scores[name]=accuracy_score(y_test,y_pred)
```

```python
for name, score in accuracy_scores.items():
  print(f'{name} accuracy: {score.mean()}')
```

```python
sns.heatmap(data=confusion_matrix(y_test, predictions['K-Nearest Neighbors']), annot=True, fmt='.0f')
```

```python
print(classification_report(y_test, predictions['K-Nearest Neighbors']))
```

```python
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler
from imblearn.combine import SMOTEENN
```

```python
def model_fit_predict():
  accuracy_scores_dict={}
  for name, model in models.items():
    model.fit(X_train_scaled, y_train)
    y_pred=model.predict(X_test_scaled)
    accuracy_scores_dict[name]=accuracy_score(y_test,y_pred)
  return accuracy_scores_dict
```

```python
over_sampling = RandomOverSampler(random_state=13)
X_os, y_os = over_sampling.fit_resample(X_train_scaled, y_train)
accuracy_scores_os=model_fit_predict()
for name, score in accuracy_scores_os.items():
  print(f'{name} accuracy: {score.mean()}')
```

```python
under_sampling = RandomUnderSampler(random_state=13)
X_us, y_us = over_sampling.fit_resample(X_train_scaled, y_train)
accuracy_scores_us=model_fit_predict()
for name, score in accuracy_scores_us.items():
  print(f'{name} accuracy: {score.mean()}')
```

```python
smote_enn = SMOTEENN(random_state=13)
X_resampled, y_resampled = smote_enn.fit_resample(X_train_scaled, y_train)
accuracy_scores_resampled={}
for name, model in models.items():
 model.fit(X_train_scaled, y_train)
 y_pred=model.predict(X_test_scaled)
 accuracy_scores_resampled[name]=accuracy_score(y_test,y_pred)
```

```python
for name, score in accuracy_scores_resampled.items():
  print(f'{name} accuracy: {score.mean()}')
```
